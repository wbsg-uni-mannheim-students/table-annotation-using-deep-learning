```python
import os
import re 
import pandas as pd
import json
import ast
directory = os.fsencode('./test_reports_final')
```

```python
result_list = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".json"): 
        with open(os.path.join(directory, file), "r") as myfile: 
            dictionary = ast.literal_eval(myfile.read())
            f1 = ("{:.2f}".format(dictionary['ts_micro_f1'] * 100))
            
        filename = filename[8:-5]

        regexp_1 = re.compile(r"(.*)_roberta-base-lr-5e-05_bs-32_ml-512_seed-(.*)_ws-(.*)_preprocess-(.*)_aug-None_mp-(.*)")
        re_match = regexp_1.match(filename)
        if (re_match): 
            list_match = list(re_match.groups())
            list_match.append(float(f1))
            result_list.append(list_match)

df = pd.DataFrame(result_list, columns=['serialization', 'seed', 'ws', 'preprocess', 'mp', 'f1'])
print(df.value_counts(['serialization', 'ws', 'preprocess', 'mp']).reset_index(name='count'))
df = df.sort_values(by=['serialization'])
df = df.groupby(['serialization', 'ws', 'preprocess', 'mp'])['f1'].mean().reset_index()
df

# # df.to_csv('results.csv', encoding='utf-8', index=False)

# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     print(df)
```
